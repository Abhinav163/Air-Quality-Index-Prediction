```python
#handling the missing values

#steps:-

#replace all the "-200" with "NaN".
#replace "NaN" with the mean of the specific column.
```

```python
#drop date and time to find mean
AirQualityData = AirQualityData.drop(['Date','Time'], axis=1)
```
